```python
import pandas as pd
```

```python

data = pd.read_excel("dds_ch2_rollingsales/rollingsales_bronx.xls", skiprows=4)

data
```

```python
columns = list(data.columns.values)
columns
```

```python
data["EASE-MENT"].sum()
```

```python
data_formated = data.drop(data.loc[data["SALE\nPRICE"] == 0].index)
data_formated.drop(["EASE-MENT"], axis=1, inplace=True)
data_formated
```

```python
def int_identify(number):
    if type(number) == type(1):
        return number
    return 1
```

```python
def year_identify(year, mean):
    if year == 0:
        return mean
    return year
```

```python
data_formated["TAX CLASS AT TIME OF SALE"] = data_formated["TAX CLASS AT TIME OF SALE"].apply(lambda x: int_identify(x))
mean_year = data_formated["YEAR BUILT"].mean()
data_formated["YEAR BUILT"] = data_formated["YEAR BUILT"].apply(lambda x: year_identify(x, mean_year))
data_formated
```

```python
grouped_data = data_formated.groupby([data_formated["SALE DATE"].dt.year, data_formated["SALE DATE"].dt.month])[["SALE\nPRICE"]].sum()
grouped_data.plot(kind = "bar", xlabel="date", ylabel="sum sale\n(1 : 100000000)", legend = "")
```

```python
grouped_data = data_formated.groupby([data_formated["SALE DATE"].dt.year, data_formated["SALE DATE"].dt.month])[["SALE\nPRICE"]].count()
grouped_data.plot(kind = "bar", xlabel="date", ylabel="number of sales")
```

```python
grouped_data = data_formated.groupby("BUILDING CLASS CATEGORY")[["SALE\nPRICE"]].count()
grouped_data.plot(kind = "bar", xlabel="Building class category", ylabel="Number of sales")
```

```python
grouped_data = data_formated.groupby("BUILDING CLASS CATEGORY")[["SALE\nPRICE"]].mean()
grouped_data.plot(kind = "bar", xlabel="Building class category", ylabel="Mean of price\n(1:1000000)")
```

```python
grouped_data = data_formated.groupby("NEIGHBORHOOD")[["SALE\nPRICE"]].count()
grouped_data.plot(kind = "bar", xlabel="NEIGHBORHOOD", ylabel="Number of sales")
```

```python
grouped_data = data_formated.groupby("YEAR BUILT")[["SALE\nPRICE"]].median()
grouped_data.plot(xlabel="YEAR BUILT", ylabel="median of price\n(1 : 10000000)", grid = "true")
```
